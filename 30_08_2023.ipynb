```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Upload your Excel file to Google Colab

# Step 2: Install necessary libraries (if not already installed)
# !pip install pandas xlrd scikit-learn matplotlib

# Step 3: Read the data from the Excel file
file_path = '/content/new.xlsx'
data = pd.read_excel(file_path)
x1 = data['x1'].values.reshape(-1, 1)
x2 = data['x2'].values.reshape(-1, 1)
x3 = data['x3'].values.reshape(-1, 1)
y = data['y'].values

# Step 4: Perform multivariate regression
X = np.concatenate((x1, x2, x3), axis=1)
model = LinearRegression()
model.fit(X, y)

# Step 5: Plot the graph
fig = plt.figure(figsize=(10, 6))

# Scatter plot of the actual data
plt.scatter(y, model.predict(X), color='blue', label='Actual vs. Predicted')

# Plot a diagonal line for reference
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linewidth=2, label='Reference Line')

plt.xlabel('Actual y')
plt.ylabel('Predicted y')
plt.title('Actual vs. Predicted y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Load the data from the Excel file
file_path = '/content/google.xlsx'  # Update with your file path
data = pd.read_excel(file_path)

# Assuming your Excel columns are named 'x1', 'x2', 'x3', and 'y'
X = data[['x1', 'x2', 'x3']]
y = data['y']

# Create a Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Generate predictions
y_pred = model.predict(X)

# Plotting in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the actual data points
ax.scatter(X['x1'], X['x2'], y, c='b', marker='o', label='Actual Data')

# Plane plot for the predicted data points
x1_range = np.linspace(X['x1'].min(), X['x1'].max(), 10)
x2_range = np.linspace(X['x2'].min(), X['x2'].max(), 10)
x1_plane, x2_plane = np.meshgrid(x1_range, x2_range)
x3_plane = model.predict(np.array([x1_plane.ravel(), x2_plane.ravel(), np.ones_like(x1_plane.ravel())]).T).reshape(x1_plane.shape)
ax.plot_surface(x1_plane, x2_plane, x3_plane, color='r', alpha=0.5, label='Regression Plane')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.show()
```
