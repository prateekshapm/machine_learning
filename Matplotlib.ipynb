```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,1,10)    #plat x from 0 to 1 with unit space or 1 differnce
y=np.sin(2*x)

fig, a=plt.subplots()
a.plot(x,y,linewidth=2.0)
a.set(xlim=(0,8),xticks=np.arange(0,8),
     ylim=(0,8),yticks=np.arange(0,8))
plt.show()
```
