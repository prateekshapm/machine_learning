## ***5.read an image***

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

image_path = '/content/Screenshot 2023-09-03 185845.png'

image = cv2.imread(image_path)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 8))
plt.imshow(image_rgb)
plt.axis('off')
plt.title('Original Image')
plt.show()

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

plt.figure(figsize=(8, 4))
plt.plot(histogram)
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

## ***7.implementation of gd***

```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    return x**2 + 3*x + 2

def gradient_function(x):
    return 2*x + 3

def gradient_descent(learning_rate, num_iterations, initial_x):
    x_history = [initial_x]
    for _ in range(num_iterations):
        gradient = gradient_function(x_history[-1])
        new_x = x_history[-1] - learning_rate * gradient
        x_history.append(new_x)
    return x_history

learning_rate = 0.1
num_iterations = 20
initial_x = -5

x_history = gradient_descent(learning_rate, num_iterations, initial_x)

plt.figure(figsize=(10, 6))
plt.plot(range(num_iterations+1), [objective_function(x) for x in x_history], marker='o', linestyle='-')
plt.xlabel('Iteration')
plt.ylabel('Objective Function Value')
plt.title('Gradient Descent Convergence')
plt.grid(True)
plt.show()

print("Optimized Parameter:", x_history[-1])
```
