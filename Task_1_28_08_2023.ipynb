2.randomly generate x and y and do linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

x=np.array([])
y=np.array([])
q=int(input("enter the number of samples you want :"))
for i in range (0,q):
  z=random.randint(1,100)
  t=random.randint(1,100)
  x=np.append(x,z)
  y=np.append(y,t)
print(y)
print(x)
x=x.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.xlabel("independent variable")
plt.ylabel("dependent variable")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
slr=LinearRegression()
slr.fit(x,y)
```

```python
pred=slr.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
print("mse: ",mse)
print("INTERCEPT :" ,slr.intercept_)
print("slpoe :",slr.coef_)
plt.plot(x,pred,color='red',marker='*')
```
